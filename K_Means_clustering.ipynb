```python
import pandas as pd
data = pd.read_csv("/content/sample_data/table.csv")
data.head()
```

```python
selected_data = data.iloc[:,0:2]
selected_data
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
k=3
kmeans = KMeans(k)
kmeans.fit(selected_data)
clusters=kmeans.fit_predict(selected_data)
clusters
```

```python
cluster_data=data.copy()
cluster_data['Clusters']=clusters
plt.scatter(cluster_data['subject-id(1-50)'],
            cluster_data['age-months'],
            c=cluster_data['Clusters'],
            cmap='rainbow')
```
