```python
import pandas as pd
dataset = pd.read_csv("/content/sample_data/students - Sheet1.csv")
dataset.head()
```

```python
dataset.info()
```

```python
male=dataset[dataset.gender=="M"]
female=dataset[dataset.gender=="F"]
```

```python
import matplotlib.pyplot as plt
plt.title("male vs female")
plt.xlabel("entry age")
plt.ylabel("hs gpa")
plt.scatter(male.entry_age,male.hs_gpa,color="red",label="male",alpha=0.3)
plt.scatter(female.entry_age,female.hs_gpa,color="blue",label="female",alpha=0.3)
plt.legend()
plt.show()
```
